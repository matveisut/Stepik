```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
df = df.rename(columns=
                                            {'parental level of education': 'parental_level_of_education',
                                            'test preparation course': 'test_preparation_course',
                                            'math score': 'math_score',
                                            'reading score': 'reading_score',
                                            'writing score': 'writing_score'})
df
```

```python
df.groupby(['gender','race/ethnicity'], as_index=False).agg({'math_score': ['mean', 'count', 'std', 'nunique'],'reading_score': ['mean', 'count', 'std']})
```

```python
df.groupby('race/ethnicity', as_index=False).agg({'math_score': ['mean', 'count', 'std'],'reading_score': ['mean', 'count', 'std']})
```
